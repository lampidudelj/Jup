```python
import matplotlib.pyplot as plt
def plot_corr(df, size=11):
    corr=df.corr()
    fig, ax=plt.subplots(figsize=(size, size))
    ax.matshow(corr)
    plt.xticks(range(len(corr.columns)), corr.columns)
    plt.yticks(range(len(corr.columns)), corr.columns)
    
plot_corr(df)
```

```python
df.corr()
```

```python
result_map={'H' : 1, 'D': 0, 'A': -1}
df['Result']= df['Result'].map(result_map)
```

```python
num_win = len(df.loc[df['Result'] == 1])
num_loose = len(df.loc[df['Result'] == -1])
num_draw = len(df.loc[df['Result'] == 0])
print("Number of Win cases: {0} ({1:3.2f}%)".format(num_win, (num_win / (num_win + num_loose + num_draw)) * 100) )
print("Number of Loss cases: {0} ({1:3.2f}%)".format(num_loose, (num_loose / (num_win + num_loose + num_draw)) * 100 ))
print("Number of Draw cases: {0} ({1:3.2f}%)".format(num_draw, (num_draw / (num_win + num_loose + num_draw)) * 100 ))
```

```python
from sklearn.cross_validation import train_test_split

feature_col_names = ['LeagueStandings', 'Powerscore', 'Rankscore', 'Odds_Home', 'Odds_Draw', 'Odds_Away']
predicted_class_names = ['Result']

x = df[feature_col_names].values
y = df[predicted_class_names].values

split_test_size = 0.30

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = split_test_size, random_state = 42) 
```

```python
print("{0:0.3f}% in training set".format((len(x_train)/len(df.index)) *100))
print("{0:0.3f}% in testing set".format((len(x_test)/len(df.index)) *100))
```

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()

nb_model.fit(x_train, y_train.ravel())
```

```python
nb_predict_train = nb_model.predict(x_train)

from sklearn import metrics

print("Accuracy: {0: .4f}".format(metrics.accuracy_score(y_train, nb_predict_train)))
print()
```

```python
nb_predict_test = nb_model.predict(x_test)

from sklearn import metrics

print("Accuracy: {0: .4f}".format(metrics.accuracy_score(y_test, nb_predict_test)))
print()
```

```python
myarray= numpy.array([[0, 0, 0, 1, 1, 1]])
colnames = ['LeagueStandings', 'Powerscore', 'Rankscore', 'Odds_Home', 'Odds_Draw', 'Odds_Away']
mydataframe = pandas.DataFrame(myarray, columns= colnames, index= [0])

nb_predict_train = nb_model.predict(mydataframe)

print(nb_predict_train)
```

```python
from sklearn.linear_model import LogisticRegressionCV
lr_cv_model = LogisticRegressionCV(n_jobs=-1, random_state=42, Cs=3, cv=10, refit=False, class_weight="balanced")  # set number of jobs to -1 which uses all cores to parallelize
lr_cv_model.fit(x_train, y_train.ravel())
```

```python
lr_cv_predict_test = lr_cv_model.predict(x_test)

# training metrics
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_test, lr_cv_predict_test)))
print(metrics.confusion_matrix(y_test, lr_cv_predict_test, labels=[1, 0, -1]) )
print("")
print("Classification Report")
print(metrics.classification_report(y_test, lr_cv_predict_test, labels=[1,0, -1]))
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42)      # Create random forest object
rf_model.fit(x_train, y_train.ravel()) 

rf_predict_train = rf_model.predict(x_train)
# training metrics
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_train, rf_predict_train)))

rf_predict_test = rf_model.predict(x_test)

# training metrics
print("Accuracy: {0:.4f}".format(metrics.accuracy_score(y_test, rf_predict_test)))

print(metrics.confusion_matrix(y_test, rf_predict_test, labels=[1, 0, -1]) )
print("")
print("Classification Report")
print(metrics.classification_report(y_test, rf_predict_test, labels=[1,0, -1]))
```

```python
from sklearn import tree
from inspect import getmembers
i_tree = 0
for tree_in_forest in rf_model.estimators_:
    with open('tree_' + str(i_tree) + '.dot', 'w') as my_file:
        my_file = tree.export_graphviz(tree_in_forest, out_file = my_file)
        print(getmembers(tree_in_forest.tree_))
    i_tree = i_tree + 1
```
